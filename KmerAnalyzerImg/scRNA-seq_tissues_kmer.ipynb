```python
%%bash
cd /input_dir/corona_analysis/input_reads/kidney_scRNA_control/
#Get app to flip bams from cellranger etc to fastq
curl -s -L http://cf.10xgenomics.com/misc/bamtofastq-1.2.0 > bamtofastq
chmod 755 bamtofastq
```
